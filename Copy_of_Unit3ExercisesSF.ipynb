## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

Describe what processes "generate" each of the previous discrete distributions models. Also report their support.

Bernouli is a probability distribution of yes or no. Yes has probability p and no has probability p-1.

Binomial is a probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials with the probability of success p.

Poisson is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time if these events occur with a known constant mean rate and independently of the time since the last event.

Geometric is a discrete probability distribution used to model the number of Bernoulli trials required to achieve the first success.

Hypergeometric is a discrete probability distribution used to model the number of successes in a sequence of draws, taken whithout replacement, from a finite population of a specific size containing a specific ampount of successful individuals.

Categorical is parameterized by a vector where each element specifies the probabilities of each possible outcome.
